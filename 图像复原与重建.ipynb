## 图像退化/复原过程的模型

在建模的时候，我们把一幅图像的退化过程简化为一个退化函数$h(x, y)$的卷积加一个加性噪声$n(x, y)$的干扰。而且这里的退化函数具有同一性，也就是对整幅图像都有相同的退化作用。

$$
g(x, y) = h(x, y)*f(x, y) + n(x, y) \Leftrightarrow G(x, y) = H(x, y)F(x, y) + N(x, y)
$$

这是一种比较理想的退化模型，虽然不能代表所有自然产生的图像退化，但还是能对图像复原方法的开发提供思路。

### 周期噪声

图像中的周期噪声是在图像获取期间由电力或者机电干扰产生的。这是本章中讨论的唯一一种空间相关噪声。这种噪声会在图像上留下一个明显的周期纹理，比如**正弦噪声**，被这种噪声污染过的图像在它们的频谱图中会有一对共轭脉冲，表示一个特定频率的波在该图像中占比较大

```python
img = cv.imread('lena.jpg', 0)
plt.figure(figsize = (12, 8))
plt.subplot(231), plt.imshow(img, cmap = 'gray'), plt.axis('off')
ft = np.log(np.fft.fftshift(np.fft.fft2(img)))
# 加入一个条状噪声
ft[:, img.shape[1] // 2] = 9
plt.subplot(232), plt.imshow(ft.real, cmap = 'gray'), plt.axis('off')
res = np.fft.ifft2(np.fft.ifftshift(np.exp(ft)))
plt.subplot(233), plt.imshow(res.real, cmap = 'gray'), plt.axis('off')
plt.subplot(234), plt.imshow(img, cmap = 'gray'), plt.axis('off')
ft = np.log(np.fft.fftshift(np.fft.fft2(img)))
# 以频域中心点开始加入一圈单位复根
central = img.shape[0] // 2 + img.shape[1] // 2 * 1j
n = 8
for i in range(n):
    t = 50 * np.exp(2j * np.pi * i / n) + central
    ft[int(t.real), int(t.imag)] = 12
res = np.fft.ifft2(np.fft.ifftshift(np.exp(ft)))
plt.subplot(235), plt.imshow(ft.real, cmap = 'gray'), plt.axis('off')
plt.subplot(236), plt.imshow(res.real, cmap = 'gray'), plt.axis('off')
plt.show()
```

这种周期噪声的去除则通常采用频率域滤波，因为这种噪声污染具有一个确定的模式，可以在频率域中很明显地看出来。

上面第二种噪声是通过加入四对共轭复根得到的，通过改变幅长和复根的中心位置，可以得到许多噪声模式（在频谱图中可以看到有一圈白点，那就是复根所在的位置）

## 只存在噪声的复原——空间滤波

第一节中介绍了图像的退化模型。如果某一个退化过程中只存在噪声的污染，那么可以选择空间滤波的方法。而空间滤波方法在前面已经详细介绍了，这里不再赘述，下面列出并实现几个教材中的典型滤波器

- 均值滤波器

    - 几何均值滤波器
    - 谐波均值滤波器
    - 逆谐波均值滤波器
- 统计排序滤波器
    - 中值滤波器
    - 最大值和最小值滤波器
    - 中点滤波器
    - 修正的阿尔法均值滤波器
- 自适应滤波器
    - 自适应局部降低噪声滤波器
    - 自适应中值滤波器
    
从前章中的空间滤波出发，编写了一个外部脚本`空间滤波器.py`，内部实现了上述几种滤波方法，下面是滤波效果的展示

```python
img = cv.imread('lena.jpg', 0) + np.random.normal(0, 10, size = img.shape)
f = space_filter(img)
plt.figure(figsize = (12, 8))
plt.subplot(231), plt.imshow(img, cmap = 'gray'), plt.axis('off') # 加噪的图像
plt.subplot(232), plt.imshow(f.func_filter(3, 'arithmetic'), cmap = 'gray'), plt.axis('off') # 算术均值滤波器
plt.subplot(233), plt.imshow(f.func_filter(3, 'geometric'), cmap = 'gray'), plt.axis('off') # 几何均值滤波器
plt.subplot(234), plt.imshow(f.func_filter(3, 'harmonic'), cmap = 'gray'), plt.axis('off') # 谐波均值滤波器
plt.subplot(235), plt.imshow(f.func_filter(3, 'inv harmonic'), cmap = 'gray'), plt.axis('off') # 逆谐波均值滤波器
plt.show()
```

上面是**均值滤波**的结果，可以看到，除了算术平均之外，其他的滤波方法都是渣渣。几何平均的一个最大的问题就是，它对0非常敏感，一旦在计算过程中遇到0则最后的结果也为0，在输出的图像上产生一块一块的斑，谐波均值滤波和逆谐波均值滤波可能有更适用的场合吧，总之在这个例子里面表现很差。

```python
img = salt_noise(cv.imread('lena.jpg', 0), 500)
f = space_filter(img)
plt.figure(figsize = (12, 8))
plt.subplot(231), plt.imshow(img, cmap = 'gray'), plt.axis('off') # 加了椒盐噪声的图
plt.subplot(232), plt.imshow(f.func_filter(3, 'median'), cmap = 'gray'), plt.axis('off') # 中值滤波器
plt.subplot(233), plt.imshow(f.func_filter(3, 'maximum'), cmap = 'gray'), plt.axis('off') # 最大值滤波器
plt.subplot(234), plt.imshow(f.func_filter(3, 'minimum'), cmap = 'gray'), plt.axis('off') # 最小值滤波器
plt.subplot(235), plt.imshow(f.func_filter(3, 'middle point'), cmap = 'gray'), plt.axis('off') # 中点滤波器
plt.subplot(236), plt.imshow(f.func_filter(3, 'alpha'), cmap = 'gray'), plt.axis('off') # 修正的阿尔法均值滤波器
plt.show()
```

上面是统计排序滤波的结果，之前我们已经知道，中值滤波器对椒盐噪声有着较好的去噪效果，相比来说，最大值和最小值滤波器适应于只有椒噪声和只有盐噪声的场合。在上面的这个例子中，修正的阿尔法均值滤波器有着更好的去噪效果。

```python
img1 = salt_noise(cv.imread('lena.jpg', 0), 500)
img2 = cv.imread('lena.jpg', 0) + np.random.normal(0, 10, size = img1.shape)
plt.figure(figsize = (12, 8))
f = space_filter(img1)
plt.subplot(231), plt.imshow(img1, cmap = 'gray'), plt.axis('off') # 加了椒盐噪声的图
plt.subplot(232), plt.imshow(f.func_filter(3, 'local auto'), cmap = 'gray'), plt.axis('off') # 自适应局部降低噪声滤波器
plt.subplot(233), plt.imshow(f.func_filter(3, 'median auto'), cmap = 'gray'), plt.axis('off') # 自适应中值滤波器
f = space_filter(img2)
plt.subplot(234), plt.imshow(img2, cmap = 'gray'), plt.axis('off') # 加了椒盐噪声的图
plt.subplot(235), plt.imshow(f.func_filter(3, 'local auto'), cmap = 'gray'), plt.axis('off') # 自适应局部降低噪声滤波器
plt.subplot(236), plt.imshow(f.func_filter(3, 'median auto'), cmap = 'gray'), plt.axis('off') # 自适应中值滤波器
plt.show()
```

上面是自适应滤波器的两种，可以看到，自适应中值滤波器对椒盐噪声的去噪效果非常好，但是至今为止都没有一个滤波器能有效地滤除高斯噪声。

自适应滤波器的想法是，噪声在整体的效果和在局部的效果是不一样的，如果能够检测一下局部的某些特性，有可能能达到一个更好的去噪效果。

## 用频率域消除周期噪声

在上一章中介绍频率域滤波的时候，提到了带通滤波器和带阻滤波器，这两个滤波器的一个重要的应用就是用来消除周期性噪声，或者提取一个噪声的模式。其中的原理非常简单：由于周期噪声一般为在频率域中分布规律的尖点造成，所以将那个尖点所在的区域乘1提取出来（带通滤波）或者乘0消除（带阻滤波）就可以获得无噪声的图像或者噪声模式。

在具体的实现中，可能不能直接乘0或乘1，为了避免上一章提到的振铃效应，需要考虑高斯滤波或者布特沃斯滤波。根据处理噪声的不同，这些滤波器分为：

1. 带阻滤波器
2. 带通滤波器
3. 陷波滤波器
4. 最佳陷波滤波器

```python
# 给图像加空间噪声
def space_noise(pic, length = 50, peak = 13, n = 8):
    ft = np.log(np.fft.fftshift(np.fft.fft2(pic)))
    central = pic.shape[0] // 2 + pic.shape[1] // 2 * 1j
    for i in range(n):
        t = length * np.exp(2j * np.pi * i / n) + central
        ft[int(t.real), int(t.imag)] = peak
    res = np.fft.ifft2(np.fft.ifftshift(np.exp(ft)))
    return res.real

# 返回图像的频谱
def frequency_domain(pic):
    return np.fft.fftshift(np.fft.fft2(pic))

# 返回频谱的空域
def space_domain(pic):
    return np.fft.ifft2(np.fft.ifftshift(pic))
```

```python
img = cv.imread('lena.jpg', 0)
noise = space_noise(img, n = 5, peak = 15)
plt.imshow(noise, cmap = 'gray'), plt.axis('off')
plt.show()
```

上面得到了一张简单的加噪图像，在分离图像的原图像和噪声模式之前，我们要定义带通滤波器和带阻滤波器的函数

1. 理想带阻滤波器
    $$
        H(u, v) = 1\quad if\ D(u, v) < D_0 - \frac W 2\\
        H(u, v) = 0\quad if\ D_0 - \frac W 2 \leq D(u, v) \leq D_0 + \frac W 2\\
        H(u, v) = 1\quad if\ D(u, v) > D_0 + \frac W 2
    $$
2. 布特沃斯带阻滤波器
    $$
        H(u, v) = \frac{1}{1 + \left(\dfrac{D(u, v)W}{D^2(u, v) - D_0^2}\right)^{2n}}
    $$
3. 高斯带阻滤波器
    $$
        H(u, v) = 1 - e^{-\dfrac 1 2\left(\dfrac{D^2(u, v) - D_0^2}{D(u, v)W}\right)^2}
    $$

```python
# 理想带阻滤波器
def IHPF(d0, w, size = (500, 500)):
    r, c, res = *size, np.ones(size, dtype = float)
    center = np.array(size) // 2
    for i in range(r):
        for j in range(c):
            t = np.linalg.norm(center - [i, j])
            if d0 - w / 2 <= t and t <= d0 + w / 2:
                res[i, j] = 0.
    return res

# 布特沃斯带阻滤波器
def BHPF(d0, w, n = 2, size = (500, 500)):
    r, c, res = *size, np.ones(size)
    center = np.array(size) // 2
    for i in range(r):
        for j in range(c):
            t = np.linalg.norm(center - [i, j])
            res[i, j] = 1 / (1 + (t * w / (t ** 2 - d0 ** 2)) ** (2 * n))
    return res

# 高斯带阻滤波器
def GHPF(d0, w, size = (500, 500)):
    r, c, res = *size, np.ones(size)
    center = np.array(size) // 2
    for i in range(r):
        for j in range(c):
            t = np.linalg.norm(center - [i, j])
            res[i, j] = 1 - np.exp(-0.5 * ((t ** 2 - d0 ** 2) / (t * w)) ** 2)
    return res
```

```python
ihpf, bhpf, ghpf = IHPF(150, 40), BHPF(150, 40), GHPF(150, 40)
```

```python
plt.figure(figsize = (12, 4))
plt.subplot(131), plt.imshow(ihpf, cmap = 'gray')
plt.subplot(132), plt.imshow(bhpf, cmap = 'gray')
plt.subplot(133), plt.imshow(ghpf, cmap = 'gray')
plt.show()
```

上面是图像，下面在三维中直观地看一下

```python
x, y = np.arange(500), np.arange(500)
xx, yy = np.meshgrid(x, y)
fig = plt.figure(figsize = (8, 12))
ax1 = fig.add_subplot(311, projection = '3d')
ax1.plot_surface(xx, yy, ihpf, cmap = 'jet')
ax2 = fig.add_subplot(312, projection = '3d')
ax2.plot_surface(xx, yy, bhpf, cmap = 'jet')
ax3 = fig.add_subplot(313, projection = '3d')
ax3.plot_surface(xx, yy, ghpf, cmap = 'jet')
plt.show()
```

现在，要对上面的噪声图像去噪了，应该做如下几步：

1. 得到加噪图像的频谱图
2. 手动调整$D_0$和$W$，使之达到最佳的去噪效果
3. 得到去噪结果

以下结果以高斯滤波为例

```python
ft = frequency_domain(noise)
ghpf = GHPF(50, 4, noise.shape)
temp = ft * ghpf
res = space_domain(temp)
noise_pattern = space_domain(ft.real * (1 - ghpf))
plt.figure(figsize = (15, 5))
plt.subplot(151), plt.imshow(noise, cmap = 'gray'), plt.axis('off')
plt.subplot(152), plt.imshow(ft.real, cmap = 'gray'), plt.axis('off')
plt.subplot(153), plt.imshow(ghpf, cmap = 'gray'), plt.axis('off')
plt.subplot(154), plt.imshow(res.real, cmap = 'gray'), plt.axis('off')
plt.subplot(155), plt.imshow(noise_pattern.real, cmap = 'gray'), plt.axis('off')
plt.show()
```

可以看到，通过这样的一个滤波的过程，我们把原图像和周期噪声给分离开了

## 估计退化函数

引起一张图像退化的原因是多种多样的，当能够正确的估计**退化函数**，就可以通过该函数得到一张高质量图像。去退化的效果好坏直接依赖于这个退化函数的准确度。

比如，在大气湍流的影响下，退化函数可以表示为

$$
    H(x, y) = e^{-k(x^2 + y^2) ^ \frac 5 6}
$$

由于运动产生的退化可以表示为

$$
    H(x, y) = \dfrac{T}{\pi(ax + by)}\sin[\pi(ax + by)]e^{-i\pi(ax + by)}
$$

```python
img = cv.imread('lena.jpg', 0)
ft = np.fft.fft2(img)
r, c = img.shape
noise1, noise2 = ft.copy(), ft.copy()
k = 0.00025
for row in range(r):
    for col in range(c):
        noise1[row, col] = np.exp(-k * (row ** 2 + col ** 2) ** (5 / 6))
T, a, b = 1, 0.1, 0.1
for row in range(r):
    for col in range(c):
        try:
            noise2[row, col] = T / (np.pi * (a * row + b * col)) * np.sin(np.pi * (a * row + b * col)) * np.exp(-1j * np.pi * (a * row + b * col))
        except:
            noise2[row, col] = 1
        
plt.figure(figsize = (12, 4))
plt.subplot(131), plt.imshow(img, cmap = 'gray'), plt.axis('off')
plt.subplot(132), plt.imshow(space_domain(ft * noise1).real, cmap = 'gray'), plt.axis('off')
plt.subplot(133), plt.imshow(space_domain(ft * noise2).real, cmap = 'gray'), plt.axis('off')
plt.show()
```
